# Data Cleaning

```python
# importing pandas
import pandas as pd
pd.set_option('display.max_columns', None)
file1 = '/content/drive/MyDrive/Colab Notebooks/c_nlp/Corona_NLP_train.csv'
file2 = '/content/drive/MyDrive/Colab Notebooks/c_nlp/Corona_NLP_test.csv'
```

```python
# reading data
# used latin1 encoding because there was an error with the encoding
df1 = pd.read_csv(file1, encoding= 'latin1')
df2 = pd.read_csv(file2, encoding= 'latin1')
```
